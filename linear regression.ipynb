```python
import pandas as pd
import numpy as np
from statsmodels.formula.api import ols
```

```python
data=pd.read_csv("movies_metadata.csv")
data["year"]=data["year"].apply(str)
data["month"]=data["month"].apply(str)
data["language"]=np.where(data['original_language'] == "en", 1, 0)
data.head()
```

```python
model= ols('np.log(revenue/1000) ~ np.log(budget/1000)+vote_average+genre+np.log(runtime)+country+year', data).fit()
model.summary2()
```

```python
results = model.summary2().as_text()
import csv
resultFile = open("table.csv",'w')
resultFile.write(results)
resultFile.close()
```
